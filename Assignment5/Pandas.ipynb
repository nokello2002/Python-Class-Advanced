```python
#checking for missing values and display in a snapshot
df.isnull().sum()
```

```python
# Add missing values using loc and iloc functions.
#1. Select the column you want to fill
voters_per_reg_centre=df['VOTERS PER REGISTRATION CENTRE']
#2. Display the top 5 elemtnts in the series
print(voters_per_reg_centre)
```

```python
#We'll impute the missing values of Voter per registration center using the mean. Here's the mean value:
voter_mean=voters_per_reg_centre.mean()
print(voter_mean)
```

```python
#With the mean, let's fill the nulls using fillna():
voters_per_reg_centre.fillna(voter_mean,inplace=True)
```
