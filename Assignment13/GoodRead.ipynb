```python
#Web scrapping with BeautifulSoup
import bs4 as bs
import urllib.request
from bs4 import BeautifulSoup
```

```python
#Connect to source data
source= urllib.request.urlopen ('https://www.goodreads.com/list/show/1.Best_Books_Ever?page=1').read()
```

```python
#Exploring the web pages and downloading them
soup = bs.BeautifulSoup(source,'lxml') 
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
#extracting all the text from a page
print(soup.get_text())
```

```python
#Parse a book page, extract book properties

#title of the page
print(soup.title)
```

```python
#get attributes
print(soup.title.name)
```

```python
# get values:
print(soup.title.string)
```

```python
# beginning navigation:
print(soup.title.parent.name)
```

```python
# getting specific values:
print(soup.p)
```

```python
print(soup.p.prettify())
```

```python
for paragraph in soup.find_all('p'):
    print(paragraph.string)
    print(str(paragraph.text))
```

```python
#Parse the page, extract book urls
for url in soup.find_all('a'):
    print(url.get('href'))
```
